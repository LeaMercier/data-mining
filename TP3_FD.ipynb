## Feature Engineering

```python
train_data = train_data.dropna()
train_data = train_data.reset_index()
# for age
for modif in range(0,len(train_data['Age'])) :
  train_data['Age'][modif] = train_data['Age'][modif] - train_data['Age'][modif]%20
# for Hours Per Week
for modif in range(0,len(train_data['HoursPerWeek'])) :
  train_data['HoursPerWeek'][modif] = train_data['HoursPerWeek'][modif] - train_data['HoursPerWeek'][modif]%20
train_data
```

We drop the attribute who seem miningless to us, which are the Income, the Income, the CapitalGain, the CapitalLoss, fnlwgt, and the education number.

```python
train_data = train_data.drop(['Income'], axis=1)
train_data = train_data.drop(['index'], axis=1)
train_data = train_data.drop(['fnlwgt'], axis=1)
train_data = train_data.drop(['EducationNum'], axis=1)
train_data = train_data.drop(['CapitalGain'], axis=1)
train_data = train_data.drop(['CapitalLoss'], axis=1)
```

We used histograms in the precedent part to see more clearly our values. So thanks to that we can guess how we are going to convert our continous values into categorical owns.

```python
def numeris (toNumerise) :
  liste = []
  liste.append(toNumerise[0])
  for i in range(0, len(toNumerise)) :
    if 11 in liste :
      toNumerise[i] = int(liste.index(toNumerise[i]))
    else :
      liste.append(toNumerise[i])
      toNumerise[i] = int(liste.index(toNumerise[i]))
  return toNumerise


# for Work class
train_data['WorkClass'] = numeris(train_data['WorkClass']).astype('int64')
# for Education
train_data['Education'] = numeris(train_data['Education']).astype('int64')
# for Marital status
train_data['MaritalStatus'] = numeris(train_data['MaritalStatus']).astype('int64')
# for Occupation
train_data['Occupation'] = numeris(train_data['Occupation']).astype('int64')
# for Relationship
train_data['Relationship'] = numeris(train_data['Relationship']).astype('int64')
# for Race
train_data['Race'] = numeris(train_data['Race']).astype('int64')
# Gender
train_data['Gender'] = numeris(train_data['Gender']).astype('int64')
# for Native country
train_data['NativeCountry'] = numeris(train_data['NativeCountry']).astype('int64')
train_data['Salary'] = train_data['Salary'].astype('int64')
```

## Model selection

Now, we are going to create the model wich would permit us to predict the differnets results

```python
from numpy import mean
from numpy import std
from sklearn.datasets import make_classification
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import RandomForestClassifier
```

```python
# creation of the dataframe
df_acc = pd.DataFrame(columns=['DTC','LR', 'GNB','RFC'])
df_sco = pd.DataFrame(columns=['DTC','LR', 'GNB','RFC'])
X = train_data.values[:, 0:9]
Y = train_data.values[:,10]

# prepare the cross-validation procedure
kf = KFold(n_splits=100)

for train_index, test_index in kf.split(X):
    
    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = Y[train_index], Y[test_index]

    # create model
    model_DTC = DecisionTreeClassifier()
    model_LR = LogisticRegression()
    model_GNB = GaussianNB()
    model_RFC = RandomForestClassifier()
    
```

## Classification on the test dataset

We train our model on all the available train dataset

```python
for train_index, test_index in kf.split(X):
    #train the model    
    model_DTC.fit(X_train, y_train)
    model_LR.fit(X_train, y_train)
    model_GNB.fit(X_train, y_train)
    model_RFC.fit(X_train, y_train)
    
    # predict the result
    y_pred_DTC = model_DTC.predict(X_test)
    y_pred_LR = model_LR.predict(X_test)
    y_pred_GNB = model_GNB.predict(X_test)
    y_pred_RFC = model_RFC.predict(X_test)
    
    # fill dataframe
    new_row = {'GNB':accuracy_score(y_test,y_pred_GNB)*100, 'DTC':accuracy_score(y_test,y_pred_DTC)*100,
               'LR':accuracy_score(y_test,y_pred_LR)*100,'RFC':accuracy_score(y_test,y_pred_RFC)*100}
    df_acc = df_acc.append(new_row, ignore_index=True)    
```

```python
print(df_acc)
print("\nAvarage accuracy of DTC :" ,df_acc['DTC'].mean())
print("Avarage accuracy of LR :" ,df_acc['LR'].mean())
print("Avarage accuracy of GNB :" ,df_acc['GNB'].mean())
print("Avarage accuracy of RFC :" ,df_acc['RFC'].mean())
```
